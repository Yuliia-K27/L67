# Основи Tensorflow

## Завдання

Створіть свій датасет на основі випадкових даних за допомогою **numpy.linspace** та проведіть на ньому лінійну регресію, використовуючи інструменти **tensorflow**.

## Розв'язок

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# Генеруємо данні
x = np.linspace(0, 50, 100)
y = np.linspace(0, 50, 100)
x += np.random.uniform(-5, 5, 100)
y += np.random.uniform(-5, 5, 100)
```

```python
# Будуємо графік
plt.scatter(x, y)
plt.show()
```

```python
# Створюємо модель нейронної мережі
model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer=tf.keras.optimizers.Adam(1e-1), loss='mse')
```

```python
# Проводимо навчання
history = model.fit(x, y, epochs=200)
```

```python
# Опис моделі
print(model.summary())
```

```python
# Робимо прогноз
y_pred = model.predict(x)
# Виводимо графік
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.show()
```
